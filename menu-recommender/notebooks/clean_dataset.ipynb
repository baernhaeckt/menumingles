### Read the dataset

```python
df_recipies = pd.read_csv("./../dataset/recipes_data.csv", low_memory=False)

print(f"Number of recipes: {len(df_recipies)}")
df_recipies.head()
```

```python
df_recipies[df_recipies["title"].str.contains("/", na=False)]
```

### Pre-processing

Pattern to replace
- W/ replace with emtpy
- /+space replace with empty
- Drop titles with urls

```python
# Replace unwanted text in title
df_recipies_clean = df_recipies.copy(deep=True)

df_recipies_clean["title"] = df_recipies_clean["title"].replace("W/", "", regex=True)
df_recipies_clean["title"] = df_recipies_clean["title"].replace("W /", "", regex=True)
df_recipies_clean["title"] = df_recipies_clean["title"].replace("/ ", "", regex=True)
df_recipies_clean["title"] = df_recipies_clean["title"].replace(r"\\", "", regex=True)
df_recipies_clean["title"] = df_recipies_clean["title"].replace(r"/M\\", "", regex=True)
df_recipies_clean["title"] = df_recipies_clean["title"].replace(r"\s+", " ", regex=True).str.strip()
df_recipies_clean["title"] = df_recipies_clean["title"].replace(r"(?<=[A-Za-z])/(?=[A-Za-z])", " / ", regex=True)

# Drop rows with URL in the title
df_recipies_clean = df_recipies_clean[~df_recipies_clean["title"].str.startswith(("HTTP", "http", "Http"), na=False)]

df_recipies_clean[df_recipies_clean["title"].str.contains("/", na=False)]
```
